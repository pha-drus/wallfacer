```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
heroescsv = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
hero_df = pd.read_csv(heroescsv)
hero_df.head()
```

## Player Count

* Display the total number of players


```python
players = len(hero_df["SN"].unique())

totalp = {"Total Players": [players]}

totalp_df = pd.DataFrame(totalp)

totalp_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(hero_df["Item Name"].unique())
avg_price = round(hero_df["Price"].mean(), 2)
purchquant = hero_df["Price"].count()
purchtot = hero_df["Price"].sum()

panalysis = {"Number of Unique Items": unique_items, 
             "Average Purchase Price": avg_price, 
             "Total Number of Purchases": purchquant, 
             "Total Revenue": purchtot}

pa_df = pd.DataFrame.from_dict(panalysis, orient="index", columns=["Purchasing Analysis"]).round(2)
pa_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
ggroup = hero_df.groupby(["Gender"])
gcount = ggroup["SN"].nunique()
gperc = round((gcount / players * 100), 2)

ganalysis = {"Total Count": gcount, 
             "Percentage of Players": gperc.astype(str) + "%"}
gd_df = pd.DataFrame.from_dict(ganalysis, orient="columns",)
gd_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gpcount = ggroup["Purchase ID"].count()
gpprice = round((ggroup["Price"].mean()), 2)
gptotal = round((ggroup["Price"].sum()), 2)
gpavg = round((gptotal / gcount), 2)

gpanalysis = {"Purchase Count": gpcount,
              "Mean Purchase Price": "$" + gpprice.astype(str),
              "Total Purchase Value": "$" + gptotal.astype(str),
              "Mean Total Per Player": "$" + gpavg.astype(str)}
gp_df = pd.DataFrame.from_dict(gpanalysis, orient="columns",)
gp_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# I chose not to use the exact bins from the example because these bin divisions felt cleaner

age_bin = [0, 9, 14, 19, 24, 29, 34, 39, 99]
age_tags = ["<10", 
            "10-14", 
            "15-19", 
            "20-24", 
            "25-29", 
            "30-34", 
            "35-39", 
            "40+"]

hero_df["Age Group"] = pd.cut(hero_df["Age"], age_bin, labels=age_tags)
age_grp = hero_df.groupby(["Age Group"])

age_tot = age_grp["SN"].nunique()
age_perc = round((age_tot / players * 100), 2)

age_df = pd.DataFrame({"Total Count": age_tot,
                       "Percentage of Players": age_perc.astype(str) + "%"})
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
apcount = age_grp["Purchase ID"].count()
apprice = round((age_grp["Price"].mean()), 2)
aptotal = round((age_grp["Price"].sum()), 2)
apavg = round((aptotal / age_tot), 2)

ap_df = pd.DataFrame({"Purchase Count": apcount,
                      "Mean Purchase Price": "$" + apprice.astype(str),
                      "Total Purchase Value": "$" + aptotal.astype(str),
                      "Mean Total Per Person": "$" + apavg.astype(str)})
ap_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spend_grp = hero_df.groupby(["SN"])

spcount = spend_grp["Purchase ID"].count()
spavg = round((spend_grp["Price"].mean()), 2)
sptotal = round((spend_grp["Price"].sum()), 2)

sp_df = pd.DataFrame({"Purchase Count": spcount,
                      "Average Purchase Price": spavg,
                      "Total Purchase Value": sptotal})

sp_df = sp_df.sort_values(["Total Purchase Value"], ascending=False)

sp_df["Average Purchase Price"] = sp_df["Average Purchase Price"].map("${:,.2f}".format)
sp_df["Total Purchase Value"] = sp_df["Total Purchase Value"].map("${:,.2f}".format)

sp_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = hero_df[["Item ID", "Item Name", "Price"]]
items_grp = items.groupby(["Item ID", "Item Name"])

itcount = items_grp["Item ID"].count()
itavg = round((items_grp["Price"].mean()), 2)
ittotal = round((items_grp["Price"].sum()), 2)

it_df = pd.DataFrame({"Purchase Count": itcount, 
                      "Item Price": itavg, 
                      "Total Purchase Value": ittotal})

it_df = it_df.sort_values(["Purchase Count"], ascending=False)

it_df["Item Price"] = it_df["Item Price"].map("${:,.2f}".format)
it_df["Total Purchase Value"] = it_df["Total Purchase Value"].map("${:,.2f}".format)

it_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
it_df = pd.DataFrame({"Purchase Count": itcount, 
                      "Item Price": itavg, 
                      "Total Purchase Value": ittotal})

it_df = it_df.sort_values(["Total Purchase Value"], ascending=False)

it_df["Item Price"] = it_df["Item Price"].map("${:,.2f}".format)
it_df["Total Purchase Value"] = it_df["Total Purchase Value"].map("${:,.2f}".format)

it_df.head()
```
