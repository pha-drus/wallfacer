```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
cdf = pd.merge(bitcoin_df, dash_df, on="Date")
cdf.head()
```

```python
# Rename columns so that they are differentiated
cdf = cdf.rename(columns={"Open_x": "BITCOIN Open", "High_x": "BITCOIN High", "Low_x": "BITCOIN Low",
                                      "Close_x": "BITCOIN Close", "Volume_x": "BITCOIN Volume", "Market Cap_x": "BITCOIN Market Cap"})

cdf = cdf.rename(columns={"Open_y": "DASH Open", "High_y": "Dash High", "Low_y": "DASH Low",
                                      "Close_y": "DASH Close", "Volume_y": "DASH Volume", "Market Cap_y": "DASH Market Cap"})

cdf.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
c_df = pd.merge(bitcoin_df, dash_df, on="Date", suffixes=("-Bitcoin", "-Dash"))
c_df.head()
```

```python
# Collecting best open for Bitcoin and Dash

bestopen_b = c_df["Open-Bitcoin"].max()
bestopen_d = c_df["Open-Dash"].max()

# Collecting best close for Bitcoin and Dash

bestclose_b = c_df["Close-Bitcoin"].max()
bestclose_d =c_df["Close-Dash"].max()

# Collecting the total volume for Bitcoin and Dash

vol_b = round(c_df["Volume-Bitcoin"].sum())
vol_d = round(c_df["Volume-Dash"].sum()              )
```
